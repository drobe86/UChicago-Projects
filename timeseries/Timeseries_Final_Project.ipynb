```python
taxi_data_2020['DateTime'] = pd.to_datetime(taxi_data_2020['Trip Start Timestamp'], format='%m/%d/%Y %I:%M:%S %p')
taxi_data_2019['DateTime'] = pd.to_datetime(taxi_data_2019['Trip Start Timestamp'], format='%m/%d/%Y %I:%M:%S %p')
taxi_data_2018['DateTime'] = pd.to_datetime(taxi_data_2018['Trip Start Timestamp'], format='%m/%d/%Y %I:%M:%S %p')
taxi_data_2017['DateTime'] = pd.to_datetime(taxi_data_2017['Trip Start Timestamp'], format='%m/%d/%Y %I:%M:%S %p')
```
